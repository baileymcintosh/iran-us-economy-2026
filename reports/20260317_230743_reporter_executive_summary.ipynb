## One-Line Status

## Situation Overview

What began as a calculated military strike — approximately 900 coordinated U.S.-Israeli sorties in 12 hours on February 28 — has cascaded into a systemic shock to the global energy infrastructure. Iran's retaliatory response targeted Persian Gulf shipping lanes, U.S. embassies, and energy facilities across the Gulf Cooperation Council (GCC). By March 4, Iran formally declared the Strait of Hormuz "closed," and within 48 hours the world's four largest container carriers — Maersk, MSC, CMA CGM, and Hapag-Lloyd — had suspended all transits.

**The quantitative picture is stark.** The Strait normally carries approximately 21 million barrels per day (mb/d) of crude, condensates, and refined products — roughly 20% of global consumption and 25–30% of all seaborne oil trade. The IEA confirmed on March 11 that Hormuz export volumes had fallen to **less than 10% of pre-conflict levels**. This is not a price shock driven by fear alone; it is a physical removal of supply at a scale the modern oil market has never absorbed.

**The qualitative picture adds critical context.** The initial strike eliminated Supreme Leader Ali Khamenei and dozens of senior officials — a decapitation of command authority that, paradoxically, *complicated* rather than accelerated any ceasefire, since Iran's retaliatory apparatus is distributed across the IRGC and regional proxies. As CBS News' veteran Middle East correspondent noted on March 13, "the U.S. and Israel underestimated Iran's capabilities, and the chance of regime collapse is very slim." President Trump stated "there is practically nothing left to target in Iran" while simultaneously deploying an additional 2,500 Marines — a rhetorical contradiction that has kept risk premiums elevated rather than easing them.

## Historical Precedents & Lessons

Three prior supply shocks offer instructive — and sobering — parallels:

**1973 Arab Oil Embargo:** A politically motivated production cut of ~4.4 mb/d quadrupled prices from $3 to $12/bbl in three months. The 2026 disruption is **4.5 times larger** at ~20 mb/d, according to FreightWaves. Critically, the 1973 embargo was a *political choice* — reversible by political decision. The 2026 Hormuz closure is a *military and physical* interdiction, requiring kinetic resolution or de-escalation of a live war to reverse.

**1979 Iranian Revolution / Second Oil Shock:** Iranian output fell 4.8 mb/d (7% of world supply). Fear and speculative hoarding caused a near-100% price rise. J.P. Morgan's research notes that across eight historical regime-change events in major oil producers, prices have spiked an average of 76% from onset to peak — and Iranian crude production has *still never recovered* to pre-1979 levels. This is the most directly analogous scenario for S4.

**1990–91 Gulf War:** Iraq's invasion of Kuwait doubled prices from $34 to $77/bbl in three months before military resolution restored stability within months. This is the closest structural analog to S1 — a decisive U.S.-led military campaign that resolves quickly. The key difference: the U.S. was fighting *to reopen* Gulf access in 1991; today the U.S. strikes are part of the *cause* of the closure.

> **🔴 Cross-Agent Insight #3 — "Law of Supply Disruption Overpricing":** The quant data pattern — a 95% price rise over 70 days followed by extreme intraday volatility — is consistent with what Capital Street FX identifies as a universal historical "law": *in every major supply shock, prices overshoot because markets price fear, not just barrels.* The qual thread adds the critical nuance: this time, the buffers are larger (IEA reserves didn't exist in 1973; U.S. shale gives the world's largest oil producer domestic insulation) but so is global absolute dependency on Hormuz. The net result, per the NYT's Bob McNally (Rapidan Energy Group), is that "the hit to supply is far greater now" than 1973, even accounting for improved policy tools.
