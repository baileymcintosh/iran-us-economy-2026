## Historical Precedents & Lessons

**1973 Arab Oil Embargo (most cited, least analogous):**
The 1973 embargo removed **~4 mb/d (7% of consumption)** via a deliberate political act. The current crisis removes **~20 mb/d (20% of consumption)** via physical military blockade. The NYT (March 13, 2026) and CFR both note today's crisis is *structurally larger*. However, the Kuwait Times (March 2026) makes the critical distinction: 1973 was a *supply refusal*; 2026 is a *physical barrier*. Military resolutions can reopen physical barriers faster than diplomatic ones can lift embargoes — which is the market's current bet. The 1973 embargo triggered a **40% S&P 500 decline** and years of stagflation. That outcome is not yet priced in.

**1990–91 Gulf War (most analogous):**
Iraq's invasion of Kuwait removed ~4.3 mb/d (roughly 7%). WTI doubled from ~$15 to $30+ in weeks. The S&P 500 fell **~20%** from peak (July 1990) to trough (October 1990) before recovering sharply when coalition forces secured the region. The Berkshire Edge analysis (March 2026) explicitly models 2026 as a "1990 template" scenario — a moderate recession risk, not a 1970s-style decade of pain. Key lesson from 1990: **the stock market bottomed *before* the war ended**, once oil prices stabilised.

**2022 Russia-Ukraine (most recent):**
IEA released ~200 million barrels (post-Ukraine). S&P 500 fell ~25%. Fed hiked 525bps. Recovery took 12 months from trough. The current IEA release of 400 million barrels is **double** the 2022 action, reflecting double the perceived severity. BMO Economics (March 2026) notes that WTI is now up roughly **$50/bbl from year-ago levels** — placing this shock "in a similar range to the five big shocks of the past 50 years."

**Critical lesson from all three:** Oil shocks cause recessions primarily when they are **prolonged beyond 90 days**. The market's 2027 futures curve pricing suggests a belief the crisis resolves within that window. If it does not, portfolio positioning needs a fundamental reset.
